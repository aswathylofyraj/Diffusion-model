```python
#---------------------import the required packages---------------------
import torch
from diffusers import StableDiffusionPipeline
from datasets import load_dataset
from PIL import Image
```

```python
#Load the model  (using a public model)
model_id = "CompVis/stable-diffusion-v1-4"
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe = pipe.to("cuda")  # Move the model to the GPU
```

```python
# function to generate images
def generate_image(prompt):
    with torch.no_grad():
        image = pipe(prompt).images[0]  # Generate the image
    return image

# promt to generate image
generated_image = generate_image(" norway witnessing northern lights")
display(generated_image)
```
